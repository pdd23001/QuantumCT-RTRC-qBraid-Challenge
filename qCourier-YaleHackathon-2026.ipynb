## Why does this matter? 

The Vehicle Routing Problem (VRP) is a combinatorial optimization problem that determines the most efficient routes for a fleet of vehicles to deliver goods or services to a set of customers, starting and ending at a central depot. The goal of the problem is to minimize total operational costs&mdash;such as distance, time, or fuel&mdash;while respecting constraints like vehicle capacity and delivery time windows. 

VRPs arise in many applications such as: 

- logistics & distribution (planning routes from a central depot to customers)
- e-commerce and last-mile delivery (managing high-volume and dynamic courier routes often with strict time windows for customer satisfaction)
- food delivery & distribution (managing perishable goods with tight time windows)
- public transit & ride-sharing (routing school buses for pick-up/drop-off, or coordinating ride-sharing services)
- field service management (routing technicians for repairs, installations, or maintenance)
- health services (scheduling daily visits for nurses or transporting patients to appointments)

Due to its computational complexity (NP-hard, meaning number of possible solutions increases **exponentially** with the size of the problem), the VRP rarely arises in the real-world where it can be solved by finding every possible combination. Instead, it is solved using classical heuristics to find near-optimal solutions efficiently. Quantum algorithms provide alternative heuristics for solving combinatorial optimization problems and can potentially generate better quality solutions faster than known classical heuristics. Even slight improvements can be substantial, (e.g., improvements in even 0.5% optimality for VRP can translate into millions of dollars per year in savings on logistics costs).

By joining this challenge, you're helping explore how quantum computing can be used to solve combinatorial problems like the VRP more efficiently, which can have significant impact in many sectors. 

## Expectations/Requirements

### Mathematical Formulation of the Optimization Problem

In this hackathon, we will focus on capacitated vehicle routing problems (CVRPs), one of several variations of the VRP. In CVRP, given a central depot, customer locations, and a number of vehicles with limited capacity, the objective is to find the best vehicle routes from the depot to customers and back to the depot such that total distance travelled by vehicles is minimized and all customers are serviced. More formally, the problem is defined as: 

$ \text{Find } \textbf{Vehicle Routes} \text{ that } \textbf{minimize Total Distance Traveled by Vehicles} $

subject to the following **constraints**:

1. Each vehicle route must start and end at the central depot.
2. Each customer must be visited by a vehicle and by only one vehicle.
3. Total number of customers visited by each vehicle should be less than or equal to its given capacity.

The first step in the hackathon is to formulate the above stated problem as a mathematical program which involves defining decision variables and formulating the objective function (i.e., Total Distance Travelled by All Vehicles) and constraints (as listed above) in terms of those decision variables.

## Solution Format: 

- For **each CVRP instance** attempted from the above list, submit a **separate text file** with the **naming convention** `Instance{Number of Instance Here}.txt`. For example, solution of Instance 1 above will have its corresponding output in a file named `Instance1.txt`.

The format of the text file is as follows: It should list the vehicle routes as a comma-separated ordered list of `customer node#` visited in the route, starting at the depot (node# 0) and ending at the depot, with one route listed per line: `r#: 0, first customer visited, ..., last customer visited, 0`. For example, for a solution with two routes, the content of the text file would look like:

```
r1: 0, 2, 3, 0
r2: 0, 1, 4, 0 
``` 

Here `r1` is the first route which starts at depot (node#=0), then visits customer #2, followed by customer #3, then returns to the depot (node#=0). The second route `r2` has a similar interpretation. 

- Please include in your presentation a resource usage table in the following format:

| CVRP Instance # | # of Qubits | # of Gate Operations | Execution Time | 
| :-: | :-: | :-: | :-: |
| | | | | 
| | | | | 
| | | | |
